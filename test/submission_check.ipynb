```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (14, 10)
import numpy as np
import random
```

```python
with open("../data/test.csv") as f:
    test_images = [line.split(',') for line in f]
test_images = np.array(test_images[1:], dtype='float32')
    
with open("../data/submission.csv") as f:
    sub_lines_all = [line.split(",")[1].rstrip("\n") for line in f]
sub_lines = sub_lines_all[1:]
```

```python
match = [(test_images[i], sub_lines[i]) for i in xrange(len(test_images))]
```

```python
## See random 'n' digit with their labels
n = 10
random.shuffle(match)
fig = plt.figure()
for i, digit in enumerate(match[:n]):
    plt.subplot(1, n, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.title(digit[1])  ## Label
    plt.imshow(digit[0].reshape((28, 28)), cmap=plt.cm.Greys)
```
